# 损失函数---> 对应高斯分布
用于衡量预测值和真实值间的区别
## 均方损失（L2 Loss）
损失函数求导相当于反向传播，可以计算出梯度，在进行梯度下降

<br>横坐标为预测值，纵坐标为损失函数的值，
蓝色曲线表示y=0是，预测值的变化，
绿色曲线为似然函数，
橙色表示梯度

## 绝对值损失函数---> 对应拉普拉斯分布

<br>横坐标为预测值，纵坐标为损失函数的值，
蓝色曲线表示y=0时，预测值的变化，即y‘偏离y的程度
绿色曲线为似然函数，代表了y'在哪取值时，这个y'对应的参数概率是最大的，
橙色表示梯度，当y‘大于0时，其导数是1，若y’小于0，则其导数为-1

Fashion-MNIST中包含的10个类别，分别为t-shirt（T恤）、trouser（裤子）、pullover（套衫）、dress（连衣裙）、coat（外套）、sandal（凉鞋）、shirt（衬衫）、sneaker（运动鞋）、bag（包）和ankle boot（短靴）。
以下函数用于在数字标签索引及其文本名称之间进行转换。


```python
def get_fashion_mnist_labels(labels):  #@save
    """返回Fashion-MNIST数据集的文本标签"""
    text_labels = ['t-shirt', 'trouser', 'pullover', 'dress', 'coat',
                   'sandal', 'shirt', 'sneaker', 'bag', 'ankle boot']
    return [text_labels[int(i)] for i in labels]
```

我们现在可以创建一个函数来可视化这些样本。


的列数。
titles（可选）：每个图像的标题列表（如果提供，必须与 imgs 的长度相同）。
scale（可选）：用于显示图像的缩放因子。默认为1.5。
计算图形大小：

该函数基于行数、列数和 scale 因子计
算整个图形的小。
创建子图：

使用 plt.subplots 创建一个子图网格，行数和列数与参数中指定的相同，并使便您可以使用单个循环遍历子图。
遍历和显示图像：

该函数然后遍历展平后的 axes 数组和 imgs 列表。对于每个子图（ax）和相应的图像（img），执行以下操作：
如果图像是PyTorch张量（torch.is_tensor(img)），则使用 .numpy() 将张量转换为NumPy数组，并使用 ax.imshow(img.numpy()) 在子图中显示它。
如果图像不是张量（假设为PIL图像），则使用 ax.imshow(img) 在子图中显示PIL图像。
它还隐藏子图上的x轴和y轴刻度，使其看起来像图像。
添加标题（如果提供）：

如果提供了 titles 参数，函数会基于 titles 列表中提供的标题为每个子图设置标题。
返回子图：

最后，函数返回子图，这在需要对图进行进一步修改时可能会有用。

```python
def show_images(imgs, num_rows, num_cols, titles=None, scale=1.5):  #@save
    """绘制图像列表"""
    # imgs：要显示的图像列表
    # num_rows：网格中的行数
    # num_cols：网格中的列数
    
    figsize = (num_cols * scale, num_rows * scale)
    _, axes = d2l.plt.subplots(num_rows, num_cols, figsize=figsize)
    
    axes = axes.flatten()#将此数组展平为一维数组，以便使用单个循环遍历子图
    
    for i, (ax, img) in enumerate(zip(axes, imgs)):
        # enumerate()函数：enumerate()是一个Python内置函数，用于将一个可迭代对象的元素与它们的索引配对。
        #在这个例子中，它被用于迭代axes和imgs，同时获取它们的索引和元素值。
        
        # for i, (ax, img)：这是一个for循环的头部，用于遍历enumerate()函数返回的元素对。
        #i表示索引，ax表示axes中的元素，img表示imgs中的元素。在每次迭代中，这些变量都会被赋予新的值。
        
        # zip(axes, imgs)：zip()函数将两个或多个可迭代对象的元素一一配对，生成一个新的可迭代对象。
        #在这里，zip(axes, imgs)将axes和imgs中相同索引位置的元素配对在一起，以便在每次迭代中同时访问它们。
        
        if torch.is_tensor(img):
            # 图片张量
            ax.imshow(img.numpy())
        else:
            # PIL图片
            ax.imshow(img)
            
        ax.axes.get_xaxis().set_visible(False) #隐藏刻度
        ax.axes.get_yaxis().set_visible(False)
        
        if titles:
            ax.set_title(titles[i])
    return axes
```

## 整合所有组件

现在我们[**定义`load_data_fashion_mnist`函数**]，用于获取和读取Fashion-MNIST数据集。
这个函数返回训练集和验证集的数据迭代器。
此外，这个函数还接受一个可选参数`resize`，用来将图像大小调整为另一种形状。


```python
def load_data_fashion_mnist(batch_size, resize=None):  #@save
    """下载Fashion-MNIST数据集，然后将其加载到内存中"""
    trans = [transforms.ToTensor()]
    if resize:
        trans.insert(0, transforms.Resize(resize))
    trans = transforms.Compose(trans)
    mnist_train = torchvision.datasets.FashionMNIST(
        root="../data", train=True, transform=trans, download=True)
    mnist_test = torchvision.datasets.FashionMNIST(
        root="../data", train=False, transform=trans, download=True)
    return (data.DataLoader(mnist_train, batch_size, shuffle=True,
                            num_workers=get_dataloader_workers()),
            data.DataLoader(mnist_test, batch_size, shuffle=False,
                            num_workers=get_dataloader_workers()))
```

下面，我们通过指定`resize`参数来测试`load_data_fashion_mnist`函数的图像大小调整功能。
